### Поиск аномалий в одномерных данных

Необходимо найти в датасете аномалии по признаку 'Profit'. 

- Из ящика с усами возьмем список аномалий и запишим результат в переменной outliers.

- Отфильтруем исходный датафрейм функцией isin() и сохраним список объектов с аномалиями в переменной df_outliers.

- Выведим количество аномалий (уже в прекоде).

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('/datasets/sales.csv')
boxplot = plt.boxplot(df['Profit'].values)

outliers = list(boxplot["fliers"][0].get_data()[1])

df_outliers = df[df["Profit"].isin(outliers)] 

print("Количество аномалий: ", len(df_outliers))
```

Количество аномалий:  1881
